```python
import secrets


def generate_secret_key(length=32):
    """
    암호화적으로 강력한 랜덤 시크릿 키를 생성합니다.
    JWT 서명 및 세션 암호화에 적합합니다.
    """
    # hex: 16진수 문자열 (길이 * 2 만큼의 문자열 생성됨)
    # token_urlsafe: URL에 안전한 문자열
    return secrets.token_hex(length)
```

```python
key = generate_secret_key()

print("\n" + "=" * 50)
print(" [ 보안 비밀키 생성기 ]")
print("=" * 50)
print(f"\n생성된 키:\n{key}\n")
print("-" * 50)
print("위 값을 backend/.env 파일의 SECRET_KEY 부분에 복사하세요.")
print("예: SECRET_KEY=" + key)
print("=" * 50 + "\n")
```
