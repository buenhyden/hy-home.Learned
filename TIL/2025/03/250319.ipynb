```python
# RxPY를 사용한 리액티브 프로그래밍 예제
from rx import of, operators as ops
import time

# 데이터 스트림 생성
data_stream = of(1, 2, 3, 4, 5)

# 데이터 스트림 변환 및 처리
processed_stream = data_stream.pipe(
    # 각 값을 제곱
    ops.map(lambda x: x * x),
    # 짝수만 필터링
    ops.filter(lambda x: x % 2 == 0),
    # 1초 간격으로
    ops.delay(1),
)

# 스트림 구독
processed_stream.subscribe(
    on_next=lambda value: print(f"데이터 수신: {value}"),
    on_error=lambda error: print(f"에러 발생: {error}"),
    on_completed=lambda: print("스트림 완료"),
)

# 비동기 실행을 위해 잠시 대기
time.sleep(6)
```
