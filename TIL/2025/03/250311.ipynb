```python
def two_sum(arr, target):
    left = 0
    right = len(arr) - 1

    while left < right:
        current_sum = arr[left] + arr[right]

        if current_sum == target:
            return [left, right]  # 목표값과 일치하는 두 숫자의 인덱스 반환
        elif current_sum < target:
            left += 1  # 합이 작으면 왼쪽 포인터를 오른쪽으로 이동
        else:
            right -= 1  # 합이 크면 오른쪽 포인터를 왼쪽으로 이동

    return [-1, -1]  # 목표값과 일치하는 두 숫자가 없는 경우


# 예시
arr = [1, 3, 4, 5, 7, 11]
target = 9
print(two_sum(arr, target))  # 출력: [1, 4] (3 + 6 = 9)
```

```python
def remove_duplicates(nums):
    if not nums:
        return 0

    # 첫 번째 포인터(slow)는 고유 요소의 위치를 추적
    # 두 번째 포인터(fast)는 배열을 순회
    slow = 0

    for fast in range(1, len(nums)):
        # 새로운 고유 요소를 발견하면
        if nums[fast] != nums[slow]:
            slow += 1  # slow 포인터 이동
            nums[slow] = nums[fast]  # 새 고유 요소를 slow 위치에 배치

    return slow + 1  # 중복이 제거된 배열의 새 길이 (인덱스 + 1)


# 예시
nums = [1, 1, 2, 2, 3, 4, 4, 5]
new_length = remove_duplicates(nums)
print(new_length)  # 출력: 5
print(nums[:new_length])  # 출력: [1, 2, 3, 4, 5]
```

```python
def reverse_bits(n, bit_count=32):
    """
    n의 비트를 뒤집음 (bit_count 비트 기준)
    """
    result = 0
    for i in range(bit_count):
        print(n, i, result)
        result = (result << 1) | (n & 1)
        n >>= 1
        print(n, i, result)
        print()
    return result
```

```python
def twoSum(nums, target):
    num_map = {}  # 딕셔너리 (해시맵) 선언

    for i, num in enumerate(nums):
        complement = target - num  # 목표 값에서 현재 숫자를 뺀 값
        print(num_map, i, num, complement)
        if complement in num_map:  # 해시맵에서 찾기
            return [num_map[complement], i]

        num_map[num] = i  # 현재 숫자 저장

    return []  # 결과가 없을 경우 (문제에서 항상 정답이 존재한다고 가정)
```

```python
def longestPalindrome(s: str) -> str:
    n = len(s)
    if n < 2:
        return s

    dp = [[False] * n for _ in range(n)]
    start, max_length = 0, 1  # 가장 긴 팰린드롬의 시작 위치와 길이
    # 길이 1짜리 문자열은 항상 팰린드롬
    for i in range(n):
        dp[i][i] = True

    print(dp)

    # 길이 2 이상 문자열 검사
    for j in range(1, n):
        for i in range(j):
            print(j, i, s[i], s[j])
            if s[i] == s[j] and (j - i <= 2 or dp[i + 1][j - 1]):
                dp[i][j] = True
                if j - i + 1 > max_length:  # 더 긴 팰린드롬이면 갱신
                    start, max_length = i, j - i + 1
    print(dp)
    return s[start : start + max_length]
```

```python
def rotate(matrix):
    n = len(matrix)

    # Step 1: 전치 행렬 만들기 (Transpose)
    for i in range(n):
        for j in range(i, n):
            matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]
    print(matrix)
    # Step 2: 각 행을 뒤집기 (Reverse Rows)
    for row in matrix:
        row.reverse()


matrix = matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
rotate(matrix=matrix)
matrix
```

```python
%%time


def solution(n):
    if n <= 0:
        return 0

    # 이진 펠린드롬의 합
    total = 0

    # 1비트 펠린드롬: 1
    if 1 <= n:
        total += 1

    # 2비트 펠린드롬: 3
    if 3 <= n:
        total += 3

    # 3비트 펠린드롬: 5, 7
    if 5 <= n:
        total += 5
    if 7 <= n:
        total += 7

    # 나머지 비트 길이의 펠린드롬
    bit_length = 4
    while True:
        # 현재 길이의 최소값이 n을 초과하면 종료
        min_value = 1 << (bit_length - 1)
        print(min_value)
        if min_value > n:
            break

        # 길이의 절반
        half_length = bit_length // 2

        # 첫 비트는 항상 1
        start = 1 << (half_length - 1)
        end = 1 << half_length

        # 홀수 길이
        if bit_length % 2 == 1:
            for i in range(start, end):
                bin_half = bin(i)[2:].zfill(half_length)

                # 중간 비트 0인 경우
                palindrome = int(bin_half + "0" + bin_half[::-1], 2)
                if palindrome <= n:
                    total += palindrome
                else:
                    return total

                # 중간 비트 1인 경우
                palindrome = int(bin_half + "1" + bin_half[::-1], 2)
                if palindrome <= n:
                    total += palindrome
                else:
                    return total

        # 짝수 길이
        else:
            for i in range(start, end):
                bin_half = bin(i)[2:].zfill(half_length)
                palindrome = int(bin_half + bin_half[::-1], 2)
                if palindrome <= n:
                    total += palindrome
                else:
                    return total

        bit_length += 1

    return total


solution(1073741823)
```
