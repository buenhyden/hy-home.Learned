```python
def heapify(arr, n, i):
    """
    주어진 배열 arr에서 인덱스 i를 루트로 하는 부분 트리를 최대 힙으로 만듭니다.
    n은 배열의 크기입니다.
    """
    largest = i  # 루트를 가장 큰 값으로 초기화
    left = 2 * i + 1  # 왼쪽 자식
    right = 2 * i + 2  # 오른쪽 자식

    # 왼쪽 자식이 루트보다 크면 largest 갱신
    if left < n and arr[left] > arr[largest]:
        largest = left

    # 오른쪽 자식이 현재까지의 largest보다 크면 갱신
    if right < n and arr[right] > arr[largest]:
        largest = right
    print(arr)
    # largest가 루트가 아니면 교환하고 재귀적으로 heapify 호출
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)


def heap_sort(arr):
    """
    힙 정렬을 사용하여 배열을 정렬합니다.
    """
    n = len(arr)

    # 배열을 최대 힙으로 구성
    # 마지막 비단말 노드부터 루트까지 거꾸로 진행
    for i in range(n // 2 - 1, -1, -1):
        print(i)
        heapify(arr, n, i)

    # 힙에서 요소를 하나씩 추출
    for i in range(n - 1, 0, -1):
        # 현재 루트(최댓값)를 마지막 요소와 교환
        arr[i], arr[0] = arr[0], arr[i]

        # 크기가 줄어든 힙에 대해 heapify 수행
        heapify(arr, i, 0)

    return arr


heap_sort([6, 9, 32, 4, 6, 34562, 10])
```

```python
def bubble_sort(arr):
    n = len(arr)

    # 배열의 모든 요소를 순회
    for i in range(n):
        # 이미 정렬된 요소는 다시 비교하지 않기 위해 n-i-1까지만 반복
        for j in range(0, n - i - 1):
            # 인접한 요소 비교 및 교환
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
            print(n, i, j, arr, arr[j], arr[j + 1])

    return arr


bubble_sort([6, 9, 32, 4, 6, 34562, 10])
```

```python
def isValid(s: str) -> bool:
    stack = []
    matching = {")": "(", "}": "{", "]": "["}

    for char in s:
        print(char, stack)
        if char in matching:  # 닫힌 괄호일 경우
            if not stack or stack.pop() != matching[char]:
                return False
        else:  # 열린 괄호일 경우
            stack.append(char)

    return not stack  # 스택이 비어 있으면 True


isValid("([])")
```

```python
def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1

        # key보다 큰 요소들을 오른쪽으로 이동
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        print(key, arr)
        # key를 올바른 위치에 삽입
        arr[j + 1] = key
        print(j, arr)

    return arr


insertion_sort([8, 5, 2, 6, 9, 3, 1, 4, 7])
```

```python
def merge_sort(arr):
    print("init", arr)
    if len(arr) <= 1:
        return arr
    # 분할 단계
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    print("ml", left)
    right = merge_sort(arr[mid:])
    print("mr", right)
    # 병합 단계
    print("a")
    return merge(left, right)


def merge(left, right):
    print(left, right)
    result = []
    i = j = 0

    # 두 배열을 비교하여 작은 값부터 결과에 추가
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    # 남은 요소들 추가

    result.extend(left[i:])
    result.extend(right[j:])
    print("result", result)
    return result


merge_sort([1, 53, 3214, 31, 34])
```

```python
def solution(v):
    # x, y 좌표 각각에 대해 XOR 연산을 사용하여 해결
    x = v[0][0] ^ v[1][0] ^ v[2][0]
    y = v[0][1] ^ v[1][1] ^ v[2][1]

    return [x, y]


# 테스트 케이스
print(solution([[1, 4], [3, 4], [3, 10]]))  # [1, 10]
print(solution([[1, 1], [2, 2], [1, 2]]))  # [2, 1]
```

```python
def solve_n_queens(n):
    solutions = []

    # 상태: 각 행별 퀸의 열 위치를 저장하는 배열
    # board[i] = j는 i번째 행의 j번째 열에 퀸이 있음을 의미
    board = [-1] * n

    def is_valid(row, col):
        # 현재 위치(row, col)에 퀸을 놓을 수 있는지 검사
        for prev_row in range(row):
            prev_col = board[prev_row]

            # 같은 열에 있는지 검사
            if prev_col == col:
                return False

            # 대각선에 있는지 검사
            if abs(prev_row - row) == abs(prev_col - col):
                return False

        return True

    def backtrack(row):
        # 모든 행에 퀸을 배치했으면 해답 추가
        print(board)
        if row == n:
            solutions.append(board[:])
            return

        # 현재 행의 각 열에 퀸 배치 시도
        for col in range(n):
            if is_valid(row, col):  # 가지치기: 유효한 위치인지 검사
                # 상태 변경: 퀸 배치
                board[row] = col
                # 다음 행으로 재귀 호출
                backtrack(row + 1)
                # 백트래킹: 퀸 제거 (명시적으로 코드에 나타내기 위해 작성, 다음 반복에서 덮어씌워짐)
                board[row] = -1

    # 첫 번째 행부터 시작
    backtrack(0)
    return solutions


solve_n_queens(4)
```
